```python
dataset_names = [
    # "Automotive",
    # "Electronics",
    # "Office_Products",
    # "Tools_and_Home_Improvement",
    # "Cell_Phones_and_Accessories",
    # "Toys_and_Games",
    "Appliances",
    # "Musical_Instruments",
]
```

```python
def report(item):
    prompt = item.prompt
    tokens = Item.tokenizer.encode(item.prompt)
    print(prompt)
    print(tokens[-20:])
    print(Item.tokenizer.batch_decode(tokens[-20:]))
```

```python
report(items[1000])
```

```python
sample = items
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(sample, test_size=0.2, random_state=42)
print(f"Divided into a training set of {len(train):,} items and test set of {len(test):,} items")
```

```python
validation_data = train[-1200:]
train = train[:-1200]
```

```python
len(train), len(validation_data), len(test)
```

```python
train_prompts = [item.prompt for item in train]
train_prices = [item.price for item in train]
test_prompts = [item.test_prompt() for item in test]
test_prices = [item.price for item in test]
validation_prompts = [item.prompt for item in validation_data]
validation_prices = [item.price for item in validation_data]
```

```python
train_dataset = Dataset.from_dict({"text": train_prompts, "price": train_prices})
test_dataset = Dataset.from_dict({"text": test_prompts, "price": test_prices})
validation_dataset = Dataset.from_dict({"text": validation_prompts, "price": validation_prices})
dataset = DatasetDict({
    "train": train_dataset,
    "test": test_dataset,
    "validation": validation_dataset
})
```

```python
from datetime import datetime
datetime.now().strftime("%Y-%m-%d")
```

```python
DATASET_NAME = "bhavin200420/amazon-appliances-dataset" + datetime.now().strftime("%Y-%m-%d")
dataset.push_to_hub(DATASET_NAME, private=True)
```

```python
with open('train.pkl', 'wb') as file:
    pickle.dump(train, file)

with open('test.pkl', 'wb') as file:
    pickle.dump(test, file)
    
with open('validation.pkl', 'wb') as file:
    pickle.dump(validation_data, file)
```
