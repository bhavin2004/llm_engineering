```python
code = """
# main.py
import os
import pandas as pd
from sqlalchemy.orm import Session
from database import Base, engine, SessionLocal
from models import DubaiOwner
from formatted_date import excel_date_to_jsdate
from get_building_name import get_building_name

def init_db():
    Base.metadata.create_all(bind=engine)

def main():
    folder_path = "./2025/data_update"
    init_db()

    db: Session = SessionLocal()

    files = [f for f in os.listdir(folder_path) if f.endswith(".xlsx")]
    for file in files:
        print(f"Processing file: {file}")
        file_path = os.path.join(folder_path, file)

        df = pd.read_excel(file_path)
        calling_api_cnt = 0

        for _, row in df.iterrows():
            # Extract building
            building = (
                row.get("Building 1")
                or row.get("Building Name 2")
                or row.get("BuildingNameEn")
                or row.get("Project")
                or row.get("Master Project")
                or row.get("Master Project Land")
                or ""
            )

            keyword = f"{building} {row.get('Master Location') or row.get('Master Project Land')}".strip()
            result = get_building_name(keyword, file, calling_api_cnt, db)

            if not result:
                continue

            api_name, api_location, calling_api_cnt = result
            formatted_date = excel_date_to_jsdate(row.get("Date"))

            # Unit No
            unit_no = row.get("Unit No") or row.get("Building No")
            if not unit_no:
                print(f"Skipping row with missing Unit Number: {row.to_dict()}")
                continue

            try:
                owner = DubaiOwner(
                    date=formatted_date,
                    source=f"2025-{file}",
                    area=api_location or row.get("Master Location") or "",
                    project_master=row.get("Master Project") or row.get("Master Project Land"),
                    project=row.get("Project"),
                    building_name=api_name or "",
                    property_type=row.get("Property Type") or row.get("PropertyTypeEn"),
                    plot_number=row.get("LandNumber"),
                    building_number=row.get("Building No") or building,
                    unit_number=unit_no,
                    size=row.get("Plot Size"),
                    owner_name=row.get("Owner Name"),
                    phone1=row.get("Phone 1"),
                    phone2=row.get("Phone 2"),
                    email=row.get("Email"),
                    nationality=row.get("Nationality"),
                )
                db.add(owner)
                db.commit()
            except Exception as e:
                print(f"Error inserting row: {e}")
                db.rollback()

        print(f"Finished file: {file}, API calls: {calling_api_cnt}")

    db.close()
    print("All Excel files processed and stored in SQLite with SQLAlchemy!")

if __name__ == "__main__":
    main()

"""
tokens = starcoder2_tokenizer.encode(code)
for token in tokens:
  print(f"{token}={starcoder2_tokenizer.decode(token)}")
```
